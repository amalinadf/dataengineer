```python
#extract
import pandas as pd
import datetime

raw_data = {
    'transaction_id': [101, 102, 103, 104, 105],
    'product_name': ['Skincare A', 'Serum B', 'Facial Wash C', None, 'Sunscreen E'],
    'price': ['Rp 150.000', 'Rp 200.000', '120000', 'Rp 50.000', 'Gratis'],
    'date': ['2026-03-01', '02/03/2026', '2026-03-03', '2026-03-04', '2026-03-05']
}

df = pd.DataFrame(raw_data)
print("--- Data Mentah ---")
print(df)
```

```python
#Transform

#drop baris product_name = none
df = df.dropna(subset=['product_name'])

#bersihin kolom price ilangin Rp, titik, dan Gratis
def clean_price(price):
    if 'Gratis' in str(price):
        return 0
    return int(str(price).replace('Rp ', '').replace('.', ''))

df['price_cleaned'] = df['price'].apply(clean_price)

#format tanggal
df['date'] = pd.to_datetime(df['date'], dayfirst=True, format='mixed')

print("Data setelah dibersihkan")
print(df[['transaction_id', 'product_name','price_cleaned', 'date']])
```

```python
import sqlite3

#create connection to DB
conn = sqlite3.connect('office_data.db')

#insert to table 'clean_transactions'
df.to_sql('clean_transactions', con=conn, if_exists='replace', index=False)

#close connection
conn.close()

print("\n[SUCCESS] data sudah masuk ke database!")
```
